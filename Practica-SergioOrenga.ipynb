Existen varios Barrios que todas sus entradas de Square.Meters son NA, vamos a eliminar del dataset todos los pisos que pertenecen a estos barrios.

```r
library(dplyr)
barrio <- df_madrid %>% group_by(Neighbourhood) %>% summarise(num_no_na=length(!na.omit(Square.Meters)))
#Nos quedamos con los barrios que tienen algún valor diferente a NA en Square.Meters
barrio_no_na <- subset(barrio, num_no_na > 0)
barrio_no_eliminar <- as.list(barrio_no_na$Neighbourhood)

#filtramos el dataset de df_madrid y nos quedamos sólo con los barrios que tienen algún valor diferente a NA en Square.Meters
df_madrid <- subset(df_madrid, Neighbourhood %in% barrio_no_eliminar)
#summary(df_madrid)
#df_madrid
```

El barrio parece ser un indicador importante para los metros cuadrados de un apartamento.

Vamos a agrupar los barrios por metros cuadrados. Podemos usar una matriz de similaridad de Tukey tal y como hicimos en el curso de estadística:

```r
tky<-TukeyHSD(aov( formula=Square.Meters~Neighbourhood, data=df_madrid ))
tky.result<-data.frame(tky$Neighbourhood)
cn <-sort(unique(df_madrid$Neighbourhood))
resm <- matrix(NA, length(cn),length(cn))
rownames(resm) <- cn
colnames(resm) <- cn
resm[lower.tri(resm) ] <- round(tky.result$p.adj,4)
resm[upper.tri(resm) ] <- t(resm)[upper.tri(resm)] 
diag(resm) <- 1
library(ggplot2)
library(reshape2)
dfResm <- melt(resm)
ggplot(dfResm, aes(x=Var1, y=Var2, fill=value))+
  geom_tile(colour = "black")+
  scale_fill_gradient(low = "white",high = "steelblue")+
  ylab("Class")+xlab("Class")+theme_bw()+
  theme(axis.text.x = element_text(angle = 90, hjust = 1),legend.position="none")
```

Usando como variable de distancia: 1-resm
Dibuja un dendrograma de los diferentes barrios.


```r
#clustering
barrios_distancia <- as.dist(1 - resm)
str(barrios_distancia)
```

```r
barrios_arbol <- hclust(barrios_distancia, method="complete")
barrios_dendrograma <- as.dendrogram(barrios_arbol)
```

```r
library(dendextend)
distancia <- 0.4
plot(color_branches(barrios_dendrograma, h=distancia))
```

¿Que punto de corte sería el aconsejable?, ¿cuantos clusters aparecen?

```r
#Con una distancia h=0.4 se generan 3 clusteres (k = 3)
library(cluster)
clusters <- cutree(barrios_dendrograma, k=3)
grafica_silhouette <- silhouette(clusters, barrios_distancia)
plot(grafica_silhouette,col=1:max(clusters),border=NA)

print("El punto de corte aconsejable es 0.4, de esa forma se generan 3 clusters")
print("Como se puede observar en la gráfica Silhouette, para los 3 clusters los objetos tienen valores altos, por lo que la clusterización con 3 clusters es apropiada.")
```

Vamos a crear una nueva columna en el dataframe df_madrid con un nuevo identificador marcado por los clusters obtenidos. Esta columna la llamaremos neighb_id

```r
clusters
#clusters[df_madrid$Neighbourhood]

df_madrid$neighb_id<-factor(clusters[df_madrid$Neighbourhood])
summary(df_madrid)
head(df_madrid)
#df_madrid
```

Vamos a crear dos grupos, uno test y otro train.

Tratamos de predecir los metros cuadrados en función del resto de columnas del dataframe.

```r
#Nos quedamos con las filas en las que Square.Meters tenga un valor diferente de NA
df_madrid_no_na <- df_madrid[which(df_madrid$Square.Meters != 'NA'),]

#establecemos la semilla y creamos el grupo de entrenamiento (70%) y el de test (30%)
set.seed(123)
idx<-sample(1:nrow(df_madrid_no_na),nrow(df_madrid_no_na)*0.7)
df_madrid.train<-df_madrid_no_na[idx,]
df_madrid.test <-df_madrid_no_na[-idx,]
```

```r
#Vamos probando entre las diferentes variables para encontrar las que generen mejores resultados en el modelo de regresión lineal
#Los resultados obtenidos considerando diferentes cantidades de variables son similares, por lo que se ha optado por utilizar un número reducido de variables en el modelo, sólo dos.

model_df_madrid<-lm(Square.Meters~Bedrooms+neighb_id,data=df_madrid.train)
summary(model_df_madrid)
```

```r
#Para saber si el modelo es bueno o malo le pasamos los valores predichos y los observados a la función caret
library(caret)

print("Training:")
df_madrid.train$pred <-  predict(model_df_madrid,df_madrid.train)
postResample(df_madrid.train$pred,obs = df_madrid.train$Square.Meters)
print("Testing:")
df_madrid.test$pred <-  predict(model_df_madrid,df_madrid.test)
postResample(df_madrid.test$pred,obs = df_madrid.test$Square.Meters)

print("Aparentemente los resultados obtenidos son aceptables")
```

```r
#Vamos a ver si hay outliers mediante la distancia de cooks
plot(cooks.distance(model_df_madrid))
print("Existen unos outliers que pueden afectar al resultado del modelo")
```

```r
#Vamos a eliminar los outliers y a volver a calcular el modelo
df_madrid.train_cooks <- df_madrid.train[cooks.distance(model_df_madrid)<0.5,]

model_df_madrid_cooks<-lm(Square.Meters~Bedrooms+neighb_id,data=df_madrid.train_cooks)
summary(model_df_madrid_cooks)
```

```r
#Pasamos los nuevos valores predichos y observados a la función caret

print("Training:")
df_madrid.train_cooks$pred <-  predict(model_df_madrid_cooks,df_madrid.train_cooks)
postResample(df_madrid.train_cooks$pred,obs = df_madrid.train_cooks$Square.Meters)
print("Testing:")
df_madrid.test$pred <-  predict(model_df_madrid_cooks,df_madrid.test)
postResample(df_madrid.test$pred,obs = df_madrid.test$Square.Meters)

print("Parece que los outliers no afectan mucho en el resultado en el grupo de test")
```

Mirad el histograma de los residuos sobre el conjunto de test para evaluar la calidad de vuestro modelo

```r
hist(df_madrid.test$Square.Meters - df_madrid.test$pred)
print("El resultado de los residuos no muestra nada extraño, por lo que parece correcto")
```

Usar PCA para encontrar el apartamento más cercano a uno dado.

Este algoritmo nos ayudaría a dado un apartamento que el algoritmo nos devolvería los 5 apartamentos más similares.


Crearemos una función tal que le pasemos un apartamento con los siguientes datos:
* Accommodates	
* Bathrooms	
* Bedrooms	
* Beds	
* Price	
* Guests.Included	
* Extra.People	
* Review.Scores.Rating	
* Latitude	
* Longitude	
* Square.Meters

y nos devuelva los 5 más similares de:

```r
#Generamos el dataframe con los datos del apartamento que se le va a pasar a la función
Accommodates_valor <- 14
Bathrooms_valor <- 4
Bedrooms_valor <- 5
Beds_valor <- 15
Price_valor <- 150
Guests.Included_valor <- 4
Extra.People_valor <-3
Review.Scores.Rating_valor <- 80
Latitude_valor <- 40.49322
Longitude_valor <- 3.686725
Square.Meters_valor <- 150

df_apartamento <- data.frame(Accommodates=Accommodates_valor, Bathrooms=Bathrooms_valor, Bedrooms=Bedrooms_valor, Beds=Beds_valor, Price=Price_valor, Guests.Included=Guests.Included_valor, Extra.People=Extra.People_valor, Review.Scores.Rating=Review.Scores.Rating_valor, Latitude=Latitude_valor, Longitude=Longitude_valor, Square.Meters=Square.Meters_valor)

#Nos quedamos con las mismas columnas en el dataframe de los pisos de Madrid, se omiten las filas que contienen NAs porque sino da un error al aplicar PCA
df_madrid_apartamentos <- na.omit(df_madrid_na[,c("Accommodates","Bathrooms","Bedrooms","Beds","Price","Guests.Included","Extra.People","Review.Scores.Rating","Latitude","Longitude","Square.Meters")])
```

```r
#Aplicamos PCA sobre el dataframe de los apartamentos de Madrid
pr_apart <- prcomp(df_madrid_apartamentos, center=TRUE, scale.=TRUE)
str(pr_apart)
```

```r
numero_componentes_pca <- 4
apart_similares <- function(pr_apart,df_apartamento){
    df_pca <- predict(pr_apart, df_apartamento)
    df_pca <- matrix(df_pca[,1:numero_componentes_pca],nrow=1)
    df_comp <- pr_apart$x[,1:numero_componentes_pca]
    dist <- rep(NA,nrow(df_comp))
    for (i in 1:nrow(df_comp)){
        dist[i]<-sum((df_pca-df_comp[i,])^2)
    }

    indices_similares_5 <- sort.list(dist) #se ordena la lista de forma ascendente, para luego quedarnos con los cinco primeros índices
    print("Los 5 apartamentos similares al indicado son los siguientes:")
    df_madrid_apartamentos[indices_similares_5 %in% 1:5,]
}

apart_similares(pr_apart, df_apartamento)
```
